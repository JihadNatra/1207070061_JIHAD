```python
#NAMA : JIHAD NATRA SITUMORANG
#NIM  : 1207070061
#KELAS: TSEB
#=============================================================================================================
#Import Library
import matplotlib.pyplot as plt
%matplotlib inline
import cv2
from skimage import data
from skimage.io import imread
from skimage.color import rgb2gray 
import numpy as np
#Load & Plot Input Image
citra1 = imread(fname="D:\Belajar Python\Praktikum 7\SourceImage\SpgBB.jpg")
citra1 = cv2.cvtColor(citra1,cv2.COLOR_RGB2GRAY)
citra2 = imread(fname="D:\Belajar Python\Praktikum 7\SourceImage\Ptrck.jpg")
citra2 = cv2.cvtColor(citra2,cv2.COLOR_RGB2GRAY)

print('Shape citra 1 : ', citra1.shape)
print('Shape citra 1 : ', citra2.shape)

fig, axes = plt.subplots(1, 2, figsize=(10, 10))
ax = axes.ravel()

ax[0].imshow(citra1, cmap = 'gray')
ax[0].set_title("Citra 1")#memasang judul pada plot
ax[1].imshow(citra2, cmap = 'gray')
ax[1].set_title("Citra 2")#memasang judul pada plot

#Code Asal Matlab
#for baris=2 : tinggi-1
#    for kolom=2 : lebar-1
#        minPiksel = min([F(baris-1, kolom-1) F(baris-1, kolom) F(baris, kolom+1) ...
#            F(baris, kolom-1) F(baris, kolom+1) F(baris+1, kolom-1)  ...
#            F(baris+1, kolom) F(baris+1, kolom+1)]);
#        
#        maksPiksel = max([F(baris-1, kolom-1) F(baris-1, kolom) F(baris, kolom+1)    ...
#            F(baris, kolom-1) F(baris, kolom+1) F(baris+1, kolom-1)  ...
#            F(baris+1, kolom) F(baris+1, kolom+1)]);    
#            
#        if F(baris, kolom) < minPiksel
#           G(baris, kolom) = minPiksel;
#        else
#            if F(baris, kolom) > maksPiksel
#                G(baris, kolom) = maksPiksel;
#            else
#                G(baris, kolom) = F(baris, kolom);
#            end
#        end    
#    end
#end

#Menyiapkan variable output
copyCitra1 = citra1.copy()
copyCitra2 = citra2.copy()

m1,n1 = copyCitra1.shape
output1 = np.empty([m1, n1])

m2,n2 = copyCitra2.shape
output2 = np.empty([m2, n2])
print('Shape copy citra 1 : ', copyCitra1.shape)
print('Shape output citra 1 : ', output1.shape)

print('m1 : ',m1)
print('n1 : ',n1)
print()

print('Shape copy citra 2 : ', copyCitra2.shape)
print('Shape output citra 2 : ', output2.shape)
print('m2 : ',m2)
print('n2 : ',n2)
print()

#Proses Filter Batas Pada Citra Input 1
for baris in range(0, m1-1):  # Iterasi baris pada range 0 hingga m1-2
    for kolom in range(0, n1-1):  # Iterasi kolom pada range 0 hingga n1-2
        
        a1 = baris  # Menyimpan nilai baris dalam variabel a1
        b1 = kolom  # Menyimpan nilai kolom dalam variabel b1
        
        arr = np.array([copyCitra1[a1-1, b1-1], copyCitra1[a1-1, b1], copyCitra1[a1, b1+1], \
            copyCitra1[a1, b1-1], copyCitra1[a1, b1+1], copyCitra1[a1+1, b1-1],  \
            copyCitra1[a1+1, b1], copyCitra1[a1+1, b1+1]])  # Membuat array arr yang berisi nilai piksel sekitar
        
        minPiksel = np.amin(arr)  # Mencari nilai piksel minimum dalam arr
        maksPiksel = np.amax(arr)  # Mencari nilai piksel maksimum dalam arr
            
        if copyCitra1[baris, kolom] < minPiksel:  # Jika nilai piksel pada copyCitra1 kurang dari minPiksel
            output1[baris, kolom] = minPiksel  # Mengisi output1 dengan nilai minPiksel
        else:  # Jika tidak
            if copyCitra1[baris, kolom] > maksPiksel:  # Jika nilai piksel pada copyCitra1 lebih dari maksPiksel
                output1[baris, kolom] = maksPiksel  # Mengisi output1 dengan nilai maksPiksel
            else:  # Jika tidak
                output1[baris, kolom] = copyCitra1[baris, kolom]  # Mengisi output1 dengan nilai piksel pada copyCitra1


#Proses Filter Batas Pada Citra Input 2
for baris1 in range(0, m2-1):  # Iterasi baris pada range 0 hingga m2-2
    for kolom1 in range(0, n2-1):  # Iterasi kolom pada range 0 hingga n2-2
        
        a1 = baris1  # Menyimpan nilai baris dalam variabel a1
        b1 = kolom1  # Menyimpan nilai kolom dalam variabel b1
        
        arr = np.array([copyCitra2[a1-1, b1-1], copyCitra2[a1-1, b1], copyCitra2[a1, b1+1], \
            copyCitra2[a1, b1-1], copyCitra2[a1, b1+1], copyCitra2[a1+1, b1-1],  \
            copyCitra2[a1+1, b1], copyCitra2[a1+1, b1+1]])  # Membuat array arr yang berisi nilai piksel sekitar
        
        minPiksel = np.amin(arr)  # Mencari nilai piksel minimum dalam arr
        maksPiksel = np.amax(arr)  # Mencari nilai piksel maksimum dalam arr
            
        if copyCitra2[baris1, kolom1] < minPiksel:  # Jika nilai piksel pada copyCitra2 kurang dari minPiksel
            output2[baris1, kolom1] = minPiksel  # Mengisi output2 dengan nilai minPiksel
        else:  # Jika tidak
            if copyCitra2[baris1, kolom1] > maksPiksel:  # Jika nilai piksel pada copyCitra2 lebih dari maksPiksel
                output2[baris1, kolom1] = maksPiksel  # Mengisi output2 dengan nilai maksPiksel
            else:  # Jika tidak
                output2[baris1, kolom1] = copyCitra2[baris1, kolom1]  # Mengisi output2 dengan nilai piksel pada copyCitra2

#Plot Citra Input dan Output Hasil dari Filter Batas
fig, axes = plt.subplots(2, 2, figsize=(10, 10))
ax = axes.ravel()

ax[0].imshow(citra1, cmap = 'gray')
ax[0].set_title("Input Citra 1")#memasang judul pada plot

ax[1].imshow(citra2, cmap = 'gray')
ax[1].set_title("Input Citra 2")#memasang judul pada plot

ax[2].imshow(output1, cmap = 'gray')
ax[2].set_title("Output Citra 1")#memasang judul pada plot

ax[3].imshow(output2, cmap = 'gray')
ax[3].set_title("Output Citra 2")#memasang judul pada plot
```
